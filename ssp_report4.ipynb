```python
import numpy as np
import matplotlib.pyplot as plt
import scienceplots
```

```python
# シード固定
np.random.seed(0)

# 1. Generate data
N = 10000
mu1 = -1.0
mu2 = 1.0
sigma1 = 1.0
sigma2 = 0.5

x1 = np.random.normal(mu1, sigma1, N)
x2 = np.random.normal(mu2, sigma2, N)
x12 = x1 + x2

mu12 = np.mean(x12)
sigma12 = np.std(x12)
```

```python
# 2. Plot data
plt.style.use(['science', 'grid', 'no-latex'])

plt.figure(figsize=(8, 6))
plt.hist(x1, bins=50, density=True, alpha=0.5, label='x1')
plt.hist(x2, bins=50, density=True, alpha=0.5, label='x2')
plt.hist(x12, bins=50, density=True, alpha=0.5, label='x1+x2')
plt.axvline(mu1, color='k', linestyle='--')
plt.axvline(mu2, color='k', linestyle='--')
plt.axvline(mu12, color='k', linestyle='--')
plt.xlabel('x')
plt.ylabel('p(x)')
plt.legend()
plt.show()
```

```python
# mu=1.0の指数分布での中心極限定理
# 1. Generate data
N = 10000
mu = 1.0
x = [np.sum(np.random.exponential(mu, N)) for _ in range(N)]
```

```python
# 2. Plot data
plt.style.use(['science', 'grid', 'no-latex'])

plt.figure(figsize=(8, 6))
plt.hist(x, bins=50, density=True, alpha=0.5, label='x')
plt.xlabel('x')
plt.ylabel('p(x)')
plt.legend()
plt.show()
```

```python
# 2項分布での中心極限定理
# 1. Generate data
N = 10000
p = 0.5
x = [np.sum(np.random.binomial(1, p, N)) for _ in range(N)]

# 2. Plot data
plt.style.use(['science', 'grid', 'no-latex'])

plt.figure(figsize=(8, 6))
plt.hist(x, bins=50, density=True, alpha=0.5, label='x')
plt.xlabel('x')
plt.ylabel('p(x)')
plt.legend()
plt.show()
```
